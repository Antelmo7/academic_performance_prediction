```python
import pandas as pd

academic_performance = pd.read_csv('enhanced_student_habits_performance_dataset.csv')
academic_performance.head()
```

```python
# nueva caracteristica ratio de horas de estudio por tiempo de pantalla
academic_performance['study_hours_per_screen_time_hours_ratio'] = academic_performance['study_hours_per_day'] / academic_performance['screen_time']
# academic_performance['study_hours_per_social_media_hours_ratio'] = academic_performance['study_hours_per_day'] / academic_performance['social_media_hours']
# academic_performance['study_hours_per_netflix_hours_ratio'] = academic_performance['study_hours_per_day'] / academic_performance['netflix_hours']
academic_performance['study_hours_per_motivation_level_ratio'] = academic_performance['study_hours_per_day'] / academic_performance['motivation_level']

# ver resultado
academic_performance[['study_hours_per_day', 'screen_time', 'study_hours_per_screen_time_hours_ratio',
                     'motivation_level', 'study_hours_per_motivation_level_ratio']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# analisis de correlaciones entre variables originales y las nuevas variables
corr_vars = ['study_hours_per_day', 'exam_score', 'screen_time', 'motivation_level',
             'study_hours_per_motivation_level_ratio', 'study_hours_per_screen_time_hours_ratio',
             'exam_anxiety_score']
corr_matrix = academic_performance[corr_vars].corr() # calcular la matriz de correlacion ( -1 / 1 )

# mapa de calor
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlacion entre variables originales y nuevas')
plt.show()
```
