```python
import pandas as pd
```

```python
# load data
academic_performance = pd.read_csv('enhanced_student_habits_performance_dataset.csv')
academic_performance.head() # read first rows
```

```python
# review general info
academic_performance.info() # es como leer la ficha medica del dataset
```

```python
# find and handle missing data
# fill in gaps with the avg instead of deleting
academic_performance.isnull().sum()
academic_performance['study_hours_per_day'] = academic_performance['study_hours_per_day'].fillna(academic_performance['study_hours_per_day'].mean())
print('null values before imputation')
print(academic_performance.isnull().sum())
```

```python
# Delete duplicates
academic_performance.drop_duplicates(inplace=True)
print('Rows before deleting duplicates:', len(academic_performance))
academic_performance.drop_duplicates(inplace=True)
print('Rows after deleting duplicates:', len(academic_performance))
```
