```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# load dataset
academic_performance = pd.read_csv('enhanced_student_habits_performance_dataset.csv')

# show first rows
academic_performance.head(9)
```

```python
# estadisticas descriptivas generales
academic_performance.describe() # resumen rapido
```

```python
# Scatterplot de horas de estudio vs califacion examen (local)
sns.scatterplot(x='study_hours_per_day', y='exam_score', data=academic_performance)
plt.title('RelaciÃ³n entre horas de estudio diarias y puntaje de examen')
plt.show()
```

```python
# mapa de calor de correlaciÃ³n entre variables clave
plt.figure(figsize=[10,6])
sns.heatmap(academic_performance[['study_hours_per_day', 'social_media_hours', 'netflix_hours', 'sleep_hours',
                                  'exercise_frequency', 'stress_level', 'screen_time',
                                  'motivation_level', 'exam_anxiety_score', 'time_management_score', 'exam_score']].corr(), annot=True, cmap='coolwarm')
plt.title('Mapa de calor - CorrelaciÃ³n entre variables')
plt.show()
```
